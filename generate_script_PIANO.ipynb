```python
# generate submit script
submit_script_dir = script_dir
all_file = os.listdir(submit_script_dir)
run_script = open('submit.fc.sh', 'w')

for f in all_file:
    if f.endswith('.r'):
        if os.path.exists(out_dir+f[:-2]+'.rank.txt'):
            continue
        submission_Script = open(submit_script_dir+f[:-2]+'.sub', 'w')
        submission_Script.write('#!/bin/sh -l\n# FILENAME: submit.sub\n'+
                               '#SBATCH --mem=32GB\n#SBATCH --nodes=1\n#SBATCH -A standby\n#SBATCH --time=4:00:00\n'+
                               '#SBATCH --job-name '+f.split('.')[0]+'\n')
        submission_Script.write('module load r/4.3.1\n')
        submission_Script.write('Rscript '+script_dir+f+'\n')
        submission_Script.close()
        run_script.write('sbatch '+submit_script_dir+f[:-2]+'.sub\n')
run_script.close()
    
```
